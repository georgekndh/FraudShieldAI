```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    classification_report,
    roc_auc_score,
    confusion_matrix,
    precision_recall_fscore_support,
)

from lightgbm import LGBMClassifier  # switched from Logistic Regression to LightGBM

# Load data
df = pd.read_csv(r"C:\Projects\FraudShieldAI\creditcard.csv")
X = df.drop(columns=["Class"])
y = df["Class"]

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

# Scale amount/time only
num_cols = ["Amount", "Time"]
pre = ColumnTransformer(
    transformers=[("num", StandardScaler(), num_cols)],
    remainder="passthrough"
)

# LightGBM pipeline
pipe = Pipeline([
    ("pre", pre),
    ("clf", LGBMClassifier(
        class_weight="balanced",
        n_estimators=600,
        learning_rate=0.05,
        num_leaves=64,
        subsample=0.9,
        colsample_bytree=0.9,
        random_state=42
    ))
])

# Train
pipe.fit(X_train, y_train)
proba = pipe.predict_proba(X_test)[:, 1]

# Choose threshold for best F1
best_t, best_f1 = 0.5, 0.0
for t in np.linspace(0.01, 0.99, 99):
    pred = (proba >= t).astype(int)
    _, _, f1, _ = precision_recall_fscore_support(
        y_test, pred, average="binary", zero_division=0
    )
    if f1 > best_f1:
        best_f1, best_t = f1, t

pred = (proba >= best_t).astype(int)
print("ROC-AUC:", roc_auc_score(y_test, proba))
print("Best F1:", best_f1, "@ threshold", round(best_t, 3))
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred, digits=4))

feature_names = list(X.columns)  # ['Time','V1',...,'V28','Amount']

# Save model
joblib.dump({"pipeline": pipe, "threshold": float(best_t)}, r"C:\Projects\FraudShieldAI\models\fraudshield_lgbm.pkl")


# ---- Visualization: PR curve, ROC curve, Confusion Matrix ----
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, roc_curve, auc, ConfusionMatrixDisplay

# PrecisionRecall
prec, rec, thr = precision_recall_curve(y_test, proba)
pr_auc = auc(rec, prec)
plt.figure()
plt.plot(rec, prec)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title(f"PrecisionRecall Curve (AUC={pr_auc:.4f})")
plt.grid(True, linestyle=":", linewidth=0.5)
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, proba)
roc_auc = auc(fpr, tpr)
plt.figure()
plt.plot(fpr, tpr)
plt.plot([0,1],[0,1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title(f"ROC Curve (AUC={roc_auc:.4f})")
plt.grid(True, linestyle=":", linewidth=0.5)
plt.show()

# Confusion Matrix Heatmap at best_t
fig, ax = plt.subplots()
ConfusionMatrixDisplay.from_predictions(y_test, pred, ax=ax, cmap="Blues", values_format='d')
ax.set_title(f"Confusion Matrix (threshold={best_t:.2f})")
plt.show()

# ---- Threshold sweep diagnostics ----
rows = []
for t in np.linspace(0.01, 0.99, 99):
    p = (proba >= t).astype(int)
    pr, rc, f1, _ = precision_recall_fscore_support(y_test, p, average='binary', zero_division=0)
    tn, fp, fn, tp = confusion_matrix(y_test, p).ravel()
    rows.append({
        'threshold': t,
        'precision': pr,
        'recall': rc,
        'f1': f1,
        'fp': fp,
        'fn': fn,
        'tp': tp,
        'tn': tn
    })

th_df = pd.DataFrame(rows)
print(th_df.sort_values('f1', ascending=False).head(10))

# Optionally save diagnostics
metrics_bundle = {
    'best_threshold': float(best_t),
    'best_f1': float(best_f1),
    'roc_auc': float(roc_auc_score(y_test, proba)),
    'pr_auc': float(pr_auc),
}
joblib.dump(metrics_bundle, r"C:\Projects\FraudShieldAI\models\fraudshield_lgbm_metrics.pkl")


# ---- Feature Importance & Explainability ----
# Get transformed feature names from ColumnTransformer
pre = pipe.named_steps["pre"]
feat_names = pre.get_feature_names_out(input_features=X_train.columns)
clf = pipe.named_steps["clf"]

# 1) LightGBM built-in feature importance (gain)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

imp_gain = clf.booster_.feature_importance(importance_type="gain")
imp_df = pd.DataFrame({"feature": feat_names, "gain": imp_gain}).sort_values("gain", ascending=False)
print(imp_df.head(20))

plt.figure()
plt.barh(imp_df.head(20)["feature"][::-1], imp_df.head(20)["gain"][::-1])
plt.title("Top 20 Features by LightGBM Gain")
plt.tight_layout()
plt.show()

# 2) Permutation Importance on the TEST set (more robust)
from sklearn.inspection import permutation_importance
X_test_trans = pre.transform(X_test)
perm = permutation_importance(
    clf,
    X_test_trans,
    y_test,
    n_repeats=5,
    random_state=42,
    n_jobs=-1,
    scoring="f1",
)
perm_df = pd.DataFrame({
    "feature": feat_names,
    "importance": perm.importances_mean,
    "std": perm.importances_std,
}).sort_values("importance", ascending=False)
print(perm_df.head(20))

plt.figure()
plt.barh(perm_df.head(20)["feature"][::-1], perm_df.head(20)["importance"][::-1])
plt.title("Top 20 Features by Permutation Importance (F1)")
plt.tight_layout()
plt.show()

# 3) SHAP explanations (sampled for speed)
import shap
idx = np.random.RandomState(42).choice(len(X_test), size=min(3000, len(X_test)), replace=False)
X_s = X_test.iloc[idx]
X_s_trans = pre.transform(X_s)

try:
    explainer = shap.TreeExplainer(clf)
    sv = explainer(X_s_trans)
    # attach names for modern shap API
    try:
        sv.feature_names = list(feat_names)
    except Exception:
        pass
    shap.plots.bar(sv, max_display=20, show=True)
    shap.plots.beeswarm(sv, max_display=20, show=True)
except Exception as e:
    # fallback for older shap API
    explainer = shap.TreeExplainer(clf)
    shap_values = explainer.shap_values(X_s_trans)
    shap.summary_plot(shap_values, features=X_s_trans, feature_names=feat_names, plot_type="bar", show=True)
    shap.summary_plot(shap_values, features=X_s_trans, feature_names=feat_names, show=True)

# 4) SHAP dependence for Amount
amount_name = [n for n in feat_names if "Amount" in n][0]
try:
    shap.plots.scatter(sv[:, amount_name], color=sv, show=True)
except Exception:
    # fallback dependence with summary_plot
    pass


# ---- SHAP Force Plot for individual explanations ----
# This will visualize why the model flagged or cleared specific transactions
import shap
import numpy as np
import matplotlib.pyplot as plt

# Pick one fraudulent and one legitimate example for contrast
fraud_idx = np.where(y_test.values == 1)[0][0]
legit_idx = np.where(y_test.values == 0)[0][0]

X_explain = X_test.iloc[[fraud_idx, legit_idx]]
X_explain_trans = pre.transform(X_explain)

explainer = shap.TreeExplainer(clf)
shap_values = explainer(X_explain_trans)

# Display the force plot inline for both examples
print("Force plot for a FRAUD case:")
shap.plots.force(shap_values[0], matplotlib=True)
plt.show()

print("Force plot for a LEGIT case:")
shap.plots.force(shap_values[1], matplotlib=True)
plt.show()

# ---- MLflow Experiment Tracking (optional) ----
import mlflow
import mlflow.sklearn

mlflow.set_tracking_uri("file:///C:/Projects/FraudShieldAI/mlruns")
mlflow.set_experiment("FraudShieldAI_LightGBM")

with mlflow.start_run(run_name="lightgbm_fraudshield_v1"):
    mlflow.log_param("n_estimators", 600)
    mlflow.log_param("learning_rate", 0.05)
    mlflow.log_param("num_leaves", 64)
    mlflow.log_metric("roc_auc", roc_auc_score(y_test, proba))
    mlflow.log_metric("best_f1", best_f1)
    mlflow.log_metric("pr_auc", pr_auc)

    mlflow.sklearn.log_model(pipe, "model")

    # Log feature importance CSV and plots
    imp_df.to_csv("C:/Projects/FraudShieldAI/models/feature_importance_gain.csv", index=False)
    perm_df.to_csv("C:/Projects/FraudShieldAI/models/feature_importance_perm.csv", index=False)
    mlflow.log_artifact("C:/Projects/FraudShieldAI/models/feature_importance_gain.csv")
    mlflow.log_artifact("C:/Projects/FraudShieldAI/models/feature_importance_perm.csv")

    print("Model, metrics, and artifacts logged to MLflow under experiment 'FraudShieldAI_LightGBM'")


# ---- Save SHAP explanations: Waterfall PNGs + Force HTML (robust) ----
import os
import matplotlib.pyplot as plt
import shap

ART_DIR = r"C:\Projects\FraudShieldAI\models"
os.makedirs(ART_DIR, exist_ok=True)

explainer = shap.TreeExplainer(clf)

# pick one fraud & one legit from REAL test set
fraud_idx = np.where(y_test.values == 1)[0][0]
legit_idx = np.where(y_test.values == 0)[0][0]

for label, idx in [("fraud", fraud_idx), ("legit", legit_idx)]:
    x_row = X_test.iloc[[idx]]
    proba_row = float(pipe.predict_proba(x_row)[0,1])
    x_row_t = pre.transform(x_row)

    # New API: Explanation object
    exp = explainer(x_row_t)

    # 1) Waterfall PNG (most reliable static image)
    shap.plots.waterfall(exp[0], max_display=15, show=False)
    png_path = os.path.join(ART_DIR, f"shap_waterfall_{label}_{idx}_p{proba_row:.3f}.png")
    plt.gcf().set_size_inches(8, 6)
    plt.savefig(png_path, bbox_inches="tight", dpi=300)
    plt.close()

    # 2) Force plot as HTML (force is primarily an HTML/JS viz)
    # Use legacy API to produce a browser-friendly force plot
    legacy_shap = explainer.shap_values(x_row_t)
    base = explainer.expected_value if np.isscalar(explainer.expected_value) else explainer.expected_value
    force_plot = shap.force_plot(base, legacy_shap[0], matplotlib=False)
    html_path = os.path.join(ART_DIR, f"shap_force_{label}_{idx}_p{proba_row:.3f}.html")
    shap.save_html(html_path, force_plot)

    # Log both to MLflow
    mlflow.set_tracking_uri("file:///C:/Projects/FraudShieldAI/mlruns")
    mlflow.set_experiment("FraudShieldAI_LightGBM")
    with mlflow.start_run(run_name="lightgbm_explanations_fixed", nested=True):
        mlflow.log_artifact(png_path)
        mlflow.log_artifact(html_path)

print("Saved SHAP artifacts (PNG waterfalls + HTML force plots) to:", ART_DIR)


# ---- SHAP with REAL FEATURE NAMES + side-by-side report ----
# Requires you already have: pipe (fitted), pre, clf, X_test, y_test
import os, numpy as np, matplotlib.pyplot as plt, shap
from pathlib import Path

ART_DIR = Path(r"C:\Projects\FraudShieldAI\models")
ART_DIR.mkdir(parents=True, exist_ok=True)

# 1) Get real transformed feature names from the ColumnTransformer
feat_names = pre.get_feature_names_out(input_features=X_train.columns)

# 2) Helper to build an Explanation *with names*
def explain_index(idx: int):
    x_row = X_test.iloc[[idx]]
    proba = float(pipe.predict_proba(x_row)[0,1])
    x_t = pre.transform(x_row)
    explainer = shap.TreeExplainer(clf)
    exp = explainer(x_t)  # shap.Explanation
    # Ensure feature names are attached (some SHAP versions omit them)
    try:
        exp.feature_names = list(feat_names)
    except Exception:
        pass
    return exp[0], proba

# 3) Pick a real fraud and legit example
fraud_idx = int(np.where(y_test.values == 1)[0][0])
legit_idx = int(np.where(y_test.values == 0)[0][0])
exp_fraud, p_fraud = explain_index(fraud_idx)
exp_legit, p_legit = explain_index(legit_idx)

# 4) Save named WATERFALL plots
fraud_png = ART_DIR / f"shap_waterfall_fraud_{fraud_idx}_p{p_fraud:.3f}.png"
legit_png = ART_DIR / f"shap_waterfall_legit_{legit_idx}_p{p_legit:.3f}.png"

shap.plots.waterfall(exp_fraud, max_display=15, show=False)
plt.gcf().set_size_inches(8, 6)
plt.savefig(fraud_png, bbox_inches="tight", dpi=300)
plt.close()

shap.plots.waterfall(exp_legit, max_display=15, show=False)
plt.gcf().set_size_inches(8, 6)
plt.savefig(legit_png, bbox_inches="tight", dpi=300)
plt.close()

print("Saved:", fraud_png, "and", legit_png)

# 5) Simple HTML report showing them side-by-side with real names in the images
report_html = ART_DIR / "shap_report_side_by_side.html"
report_html.write_text(f"""
<!doctype html>
<meta charset='utf-8'>
<title>FraudShield AI  SHAP Report</title>
<h2>SHAP Waterfall Explanations (with real feature names)</h2>
<p><b>Fraud example</b> (idx={fraud_idx}, proba={p_fraud:.3f}) &nbsp; | &nbsp; <b>Legit example</b> (idx={legit_idx}, proba={p_legit:.3f})</p>
<div style='display:flex; gap:20px;'>
  <img src='{fraud_png.name}' style='width:48%; border:1px solid #ddd;'>
  <img src='{legit_png.name}' style='width:48%; border:1px solid #ddd;'>
</div>
<p style='margin-top:12px;color:#666'>Feature names come from the fitted ColumnTransformer: {', '.join(feat_names[:6])}</p>
""", encoding="utf-8")

print("HTML report:", report_html)

# 6) (Optional) Log to MLflow
try:
    import mlflow
    mlflow.set_tracking_uri("file:///C:/Projects/FraudShieldAI/mlruns")
    mlflow.set_experiment("FraudShieldAI_LightGBM")
    with mlflow.start_run(run_name="shap_named_report", nested=True):
        mlflow.log_artifact(str(fraud_png))
        mlflow.log_artifact(str(legit_png))
        mlflow.log_artifact(str(report_html))
except Exception as e:
    print("MLflow logging skipped:", e)
```
